```python
import pandas as pd
df= pd.read_csv('/content/BigBasket Products.csv')
df.head()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['product'].fillna('Unkown' , inplace=True)
df['brand'].fillna('Unkown', inplace = True)
df.isnull().sum()
```

```python
df['rating'].fillna(df['rating'].mean(), inplace=True)
df['rating'].isnull().sum()
```

```python
df['description'].fillna('Unkown' , inplace=True)
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.to_csv('Bigbasketcleaned_dataset.csv', index=False)
print("Cleaned dataset saved: Bigbasketcleaned_dataset.csv")
```
